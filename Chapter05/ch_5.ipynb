### Impute: KNN 

```python
from sklearn.impute import KNNImputer
```

```python
df_knn_impute = df.copy()
df_knn_impute.drop(['Major','P25th','P75th'], axis = 1, inplace = True)
```

```python
df_knn_impute.head()
```

#### One hot encoding

```python
columns_to_encode = ['Major_category']
one_hot_df = pd.get_dummies(df_knn_impute, prefix=['one_hot'], columns = columns_to_encode) 

one_hot_df
```

#### Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler() 

scaled_one_hot_df = min_max_scaler.fit_transform(one_hot_df) 
```

```python
df_scaled = pd.DataFrame(scaled_one_hot_df,columns=one_hot_df.columns)
df_scaled
```

```python
import numpy as np
imputer = KNNImputer(missing_values=np.nan, n_neighbors=3)
```

#### Imputing

```python
imputed_df = imputer.fit_transform(df_scaled)
knn_df = pd.DataFrame(imputed_df,columns=df_scaled.columns)
```

```python
knn_df.isna().any()
```

```python
knn_df['Median'].iloc[170]
```

```python
df_mean_impute.isnull().sum()
```

```python
df_knn_impute['Median'] = knn_df['Median']
df_knn_impute
```

# Matplotlib

### Group By

```python
df.groupby('Major_category')[['Major_category','Median']].median()
# df.groupby('Major_category').sum()
```

```python
import matplotlib.pyplot as plt
```

```python
df_major_cat = df.groupby('Major_category').mean()
df_major_cat
```

```python
df_major_cat = df.groupby('Major_category').agg(
    Total = ('Total', 'sum'),
    Median_Salary =  ('Median','mean'),
)
df_major_cat
```

```python
df_major_cat.sort_values(by='Median_Salary',ascending=False, inplace=True)
df_major_cat
```

```python
from matplotlib.pyplot import figure

plt.bar(df_major_cat.index,df_major_cat['Median_Salary'],color='blue', label="Major")
plt.xticks(rotation=90)

plt.xlabel('Major')
plt.ylabel("Median Salary")
plt.title('Barchart - Median Salary vs Major ',fontsize=15)
plt.legend()
plt.figure(figsize=( 6,3)  , dpi=200)
plt.show()
```

## Correlation of pay and number of people in it

```python
df_major_cat
```

```python
import numpy as np

x = df_major_cat['Total']
y = df_major_cat['Median_Salary']

plt.scatter(x,y)
plt.xlabel('Number of Students')
plt.ylabel('Median')

poly = np.polyfit(x, y, 1)

p = np.poly1d(poly)
plt.plot(x, p(x))
```

## List of Methods & their Explanations

df.groupby() - Group by an indicated column.

df.sort_values() - Specify which columns to sort the DataFrame on.

df.fillna() - Fill in all NaN values using the supplied approach.

pd.isnull(df) - Detects whether there are missing values.

df.loc['row1'] and df.iloc[1] - Search for specific element by column labels or by integer indexes respectively.

df.query('col1 > col2') - Perform a basic query with Boolean expression. In our example, it owuld return any row where the value in the first column was greater than the second column.
